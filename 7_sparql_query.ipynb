```python
from SPARQLWrapper import SPARQLWrapper, JSON

sparql = SPARQLWrapper("http://192.168.133.39:8890/sparql")

query = """
PREFIX ex:   <http://example.org/schema#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX wd:   <http://www.wikidata.org/entity/>
PREFIX wdt:  <http://www.wikidata.org/prop/direct/>

SELECT DISTINCT ?person ?localLabel
WHERE {
  ?event ex:hasTriple ?triple .
  ?triple ex:subject ?person .
  FILTER(BOUND(?person))  # 確保有 person

  OPTIONAL {
    ?person rdfs:label ?localLabel .
    FILTER(LANGMATCHES(LANG(?localLabel), "zh") || LANGMATCHES(LANG(?localLabel), "zh-TW"))
  }
}
LIMIT 100
"""

sparql.setReturnFormat(JSON)

results = sparql.query().convert()

#results



for i, result in enumerate(results["results"]["bindings"]):
    if "person" not in result:
        print(f"⚠️ 第 {i+1} 筆無 person 欄位：{result}")
    else:
        person = result["person"]["value"]
        label = result.get("localLabel", {}).get("value", "(無標籤)")
        print(f"{person} -> {label}")
        
        
# for result in results["results"]["bindings"]:
#     person = result.get("person", {}).get("value")
#     label = result.get("localLabel", {}).get("value")
    
#     if not person:
#         print("⚠️ 無 person 欄位，跳過此筆")
#         continue

#     if not label:
#         print(f"{person} 缺少中文標籤，可補查 Wikidata")
#     else:
#         print(f"{person} -> {label}")
```
